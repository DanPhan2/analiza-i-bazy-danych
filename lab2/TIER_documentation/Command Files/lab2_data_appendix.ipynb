```python
import pandas as pd

df = pd.read_fwf("../Original Data/weather.txt", widths=[11,4,2,4]+[5,3]*31, header=None)
df
```

```python
DF = df.dropna()
DF
```

```python
df = DF.drop(columns=DF.columns[(DF == 'I').any()])
df = df.reset_index(drop=True)
df
```

```python
df_new = df.melt(id_vars = [0, 1, 2, 3])
```

```python
def change_day(x):
    return round((x-2)/2)

df_new.variable = df_new.variable.apply(change_day)
df_new=df_new.rename(columns={0: "ID", 1: "YEAR", 2:"MONTH", 3:"VAR_NAME", "variable":"DAY", "value": "VALUE"})
df_new
```

```python
df_new['roboczy']=df_new['YEAR'].astype(str) + df_new['MONTH'].astype(str).str.zfill(2)+ df_new['DAY'].astype(str).str.zfill(2)
df_new['DATE'] = pd.to_datetime(df_new['roboczy'], format='%Y%m%d')
df_new=df_new.drop(["roboczy", "YEAR", "MONTH", "DAY"], axis = 1)
df_new
```

```python
clean=df_new.pivot(index='DATE', columns='VAR_NAME', values='VALUE')
clean = clean.dropna()
clean
```
