Należało zdefiniować pewną funkcję (fcn) i sporządzić dla niej wykresy z podanych przedziałów:

```python
import matplotlib.pyplot as plt
import numpy as np

def fcn(x):
    return x*x+5

def plot_fcn(xmin, xmax):
    x = np.linspace(xmin,xmax,1000)
    y = fcn(x)
    plt.plot(x,y)
    plt.title("Wykres funkcji x*x + 5 dla x > {} i x < {}".format(xmin,xmax))
    plt.legend(["fcn"])
    plt.ylabel("Y")
    plt.xlabel("X")
    plt.show()
```

Przedział (-1,1)

```python
plot_fcn(-1,1)
```

Przedział (-6,6)

```python
plot_fcn(-6,6)
```

Przedział (0,5)

```python
plot_fcn(0,5)
```
